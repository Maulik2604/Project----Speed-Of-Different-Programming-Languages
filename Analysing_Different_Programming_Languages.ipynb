```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
y  = [1, 2, 3, 4, 5]
x1 = [50, 60, 70, 80, 90]
x2 = [52, 63, 74, 85, 96]
x3 = [54, 66, 78, 90, 102]
x4 = [60, 75, 95, 120, 150]
x5 = [65, 85, 105, 135, 185]
pics = sns.lineplot(y = x1,x = y,  label='C', linewidth=2)
sns.lineplot(y = x4,x = y,  label='Perl', linewidth=2)
sns.lineplot(y = x2,x = y,  label='C++', linewidth=1)
sns.lineplot(y = x5,x = y,  label='Python', linewidth=2)
sns.lineplot(y = x3,x = y,  label='Java', linewidth=3)
pics.set_xlabel('File size(GB)')
pics.set_xlabel('Time Taken(sec)')
pics.set_title('Speed of different programming languages')
```
